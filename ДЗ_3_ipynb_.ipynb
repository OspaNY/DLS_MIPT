Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:

```python
n_cols = 4
n_rows = (len(cat_cols) + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
if n_rows == 1:
    axes = [axes] if n_cols == 1 else axes
else:
    axes = axes.flatten()

for i, col in enumerate(cat_cols):
    # Получаем value_counts и сортируем по убыванию
    value_counts = data[col].value_counts().sort_values(ascending=False)

    # Строим bar chart
    bars = axes[i].bar(value_counts.index.astype(str), value_counts.values, alpha=0.7)
    axes[i].set_title(f'{col}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Количество')

    # Добавляем подписи значений на столбцы
    for bar, count in zip(bars, value_counts.values):
        height = bar.get_height()
        axes[i].text(bar.get_x() + bar.get_width()/2., height + 0.1,
                    f'{count}', ha='center', va='bottom', fontsize=9)

    axes[i].tick_params(axis='x', rotation=-15)

    # Устанавливаем выравнивание подписей по левому краю
    for tick in axes[i].get_xticklabels():
        tick.set_horizontalalignment('left')
        tick.set_rotation_mode('anchor')  # Вращение вокруг точки выравнивания

# Скрываем пустые subplots
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data[num_cols]), columns=num_cols, index=data.index)
```

```python
data_scaled.hist()
```

```python
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_array = encoder.fit_transform(data[cat_cols])
feature_names = encoder.get_feature_names_out(cat_cols)
encoded_df = pd.DataFrame(encoded_array, columns=feature_names, index=data.index)
data_final = pd.concat([data_scaled, encoded_df], axis=1)
```

```python
data_final.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    data_final,
    data[target_col],
    test_size = 0.2,
    random_state = 42,
    stratify = data[target_col]  # Стратификация для сохранения распределения классов
)
```

```python
model = LogisticRegressionCV(
    Cs=10,  # Количество значений C для перебора (или список конкретных значений)
    cv=5,   # Количество фолдов в кросс-валидации
    scoring='roc_auc',
    refit=True,  # Переобучить на всех данных с лучшими параметрами
    random_state=42,
    max_iter=1000,
    n_jobs=-1  # Использовать все ядра процессора
)
```

```python
model.fit(X_train, y_train)
```

```python
print(f"Лучший параметр C: {model.C_[0]}")
print(f"Лучшее ROC-AUC на кросс-валидации: {model.scores_[1].mean(axis=0).max():.4f}")
```

```python
y_pred_proba = model.predict_proba(X_valid)[:, 1]
y_pred = model.predict(X_valid)

# Оценка качества на валидационной выборке
valid_roc_auc = roc_auc_score(y_valid, y_pred_proba)
print(f"ROC-AUC на валидационной выборке: {valid_roc_auc:.4f}")
```

```python
!pip install catboost -q
from catboost import CatBoostClassifier
```

```python
data_combined = pd.concat([data_scaled, data[cat_cols]], axis=1)
X_train, X_valid, y_train, y_valid = train_test_split(
    data_combined,
    data[target_col],
    test_size = 0.2,
    random_state = 42,
    stratify = data[target_col]  # Стратификация для сохранения распределения классов
)
```

```python
cat_model = CatBoostClassifier(
    cat_features=cat_cols,
    random_state=42,
    verbose=100,  # Вывод каждые 100 итераций
    early_stopping_rounds=50
)
```

```python
cat_model.fit(
    X_train, y_train,
    eval_set=(X_valid, y_valid),
    plot=True  # Можно установить True для визуализации обучения
)
```

```python
y_pred_proba_default = cat_model.predict_proba(X_valid)[:, 1]
roc_auc_default = roc_auc_score(y_valid, y_pred_proba_default)

print(f"ROC-AUC со стандартными параметрами: {roc_auc_default:.4f}")
```

```python
param_grid = {
    'n_estimators': [100, 200, 500, 1000],
    'learning_rate': [0.01, 0.05, 0.1, 0.2]
}

best_score = 0
best_params = {}

# Перебираем все комбинации параметров
for n_estimators in param_grid['n_estimators']:
    for learning_rate in param_grid['learning_rate']:
        print(f"\nТестируем: n_estimators={n_estimators}, learning_rate={learning_rate}")

        # Создаем и обучаем модель
        best_cat_model = CatBoostClassifier(
            n_estimators=n_estimators,
            learning_rate=learning_rate,
            cat_features=cat_cols,
            random_state=42,
            verbose=False  # Отключаем вывод для чистоты
        )

        best_cat_model.fit(X_train, y_train)

        # Оцениваем на валидационной выборке
        y_pred_proba = best_cat_model.predict_proba(X_valid)[:, 1]
        roc_auc = roc_auc_score(y_valid, y_pred_proba)

        print(f"ROC-AUC: {roc_auc:.4f}")

        # Сохраняем лучшую комбинацию
        if roc_auc > best_score:
            best_score = roc_auc
            best_params = {
                'n_estimators': n_estimators,
                'learning_rate': learning_rate
            }
            best_model = best_cat_model

print(f"Лучшая комбинация параметров: {best_params}")
print(f"Лучший ROC-AUC: {best_score:.4f}")
print(f"Улучшение относительно стандартных параметров: {best_score - roc_auc_default:.4f}")
```
